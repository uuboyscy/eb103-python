```python
import pymysql
```

```python
host = '127.0.0.1'
port = 3306
user = 'root'
passwd = 'root'
db = 'TESTDB'
charset = 'utf8mb4'
```

```python
conn = pymysql.connect(host=host,
                      port=port,
                      user=user,
                      passwd=passwd,
                      db=db,
                      charset=charset)
```

```python
cursor = conn.cursor()
```

```python
cursor.close()
conn.close()
```
